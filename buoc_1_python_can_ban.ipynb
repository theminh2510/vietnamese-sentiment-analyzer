```python
comments = ["san pham tot", "giao hang cham", "rat hai long", "te qua"]
print(comments[0]) #Lay phan tu dau
print(len(comments)) #Dem so phan tu
for comment_item in comments:
  print(comment_item)
comment_dict = {
    'text' : 'san pham tot lam',
    'label' : 'tich cuc'
}

print(comment_dict["text"])
print(comment_dict["label"])
def phan_loai(binh_luan):
  tu_tich_cuc = ["tot", "hai long", "tuyet", "ok"]
  tu_tieu_cuc = ["cham", "te", "khong hai long"]
  for tu in tu_tich_cuc:
    if tu in binh_luan:
      return "tich cuc"

  for tu in tu_tieu_cuc:
    if tu in binh_luan:
      return "tieu cuc"
  return "khong ro"


#Thu chay
print(phan_loai("san pham tot lam"))
print(phan_loai("giao hang cham"))
print(phan_loai("ok"))

#Ghi file
with open("comments.txt", "w",  encoding="utf-8") as f:
  f.write("san pham tot\n")
  f.write("giao hang cham\n")

#Doc file
with open("comments.txt", "r",  encoding="utf-8") as f:
  for line in f:
    print(line.strip())
```
